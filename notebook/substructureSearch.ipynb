```julia
path =  joinpath(dirname(@__FILE__), "..", "_resources", "open structures.sdf")
mols = sdfilereader(open(path))
function substrsearch(query)
    matched = []
    for (i, m) in enumerate(mols)
        if match_molquery(m, query)
            push!(matched, m)
            print("@")
        else
             print("+")
        end
        if i % 50 == 0
            println(i)
        end
    end
    println("$(length(matched)) records matched")
    return matched
end
```
